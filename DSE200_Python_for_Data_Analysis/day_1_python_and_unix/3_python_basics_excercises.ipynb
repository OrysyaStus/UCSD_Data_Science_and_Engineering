### (9) Take as an input an array of integers, and return true if the array contains a 5 immediately after another 5 somewhere.

```python
A=[1,5,1,6,5,3,5,2]
B=[1,2,3,4,5,5,6,7,8,9]
```

```python
def has55(num):
    it = iter(num)
    return any(x == 5 == next(it) for x in it)
```

<img src="8queen.jpg">

```python
from itertools import permutations

def board(vec):
    for col in vec:
        s = ['.'] * len(vec)
        s[col] = 'Q'
        print ''.join(s)
    print
    
n = 8
cols = range(n)
solutions = []
for vec in permutations(cols):
    if (n == len(set(vec[i]+i for i in cols))
          == len(set(vec[i]-i for i in cols))):
        print vec
        print 'Vectors translated to chess board' 
        board(vec)
        solutions.append(vec)
        
print 'There are', len(solutions), 'solutions in total.'
```

### (13) Write a program to make a list of all prime numbers less than 3000. A prime number can only be divided evenly by 1 and itself.

```python
primes = []
for x in range(1, 3000):
    prime = True
    for i in range(2,x):
        if (x%i==0):
            prime = False
    if prime == True:
        primes.append(x)
primes
```

### (14) Solve the above problem using Set instead of List 

```python
primes = set()
for x in range(1, 3000):
    prime = True
    for i in range(2,x):
        if (x%i==0):
            prime = False
    if prime == True:
        primes.add(x)
primes
```
